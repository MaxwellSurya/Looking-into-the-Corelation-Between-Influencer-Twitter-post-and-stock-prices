```
# A function to abstract the stock & trade data of a company for each year, month, and/or date
def stock_report(company, year, month = 0, day = 0):
    if month == 0:  
        rgx = "\d+\/\d+\/" + (str) (year % 100)
    elif day == 0:
        assert(1 <= month <= 12)
        rgx = (str) (month) + "\/\d+\/" + (str) (year % 100)
    else:
        
        return company.loc[company['Date'] == (str) (month) + "/" + (str) (day) + "/" + (str) (year % 100)]
    target = company['Date'].str.contains(rgx, regex = True)
    company['target'] = target
    company = company.loc[company['target'] == True]
    filtered_table = company.drop(columns = ['target']).reset_index().drop(columns = ['index'])
    return filtered_table
```

```
# A function to get weekly report from a particular day
# Make a stock report with data of 'company' for 'n' days from 'year'/'month'/'day'
def stock_report_after_n_days(company, n, year, month, day):
    stock_n_days = pd.DataFrame()
    stock = stock_report(company, year, month, day)
    while (stock.empty == True):
        day += 1
        stock = stock_report(company, year, month, day)
    index = company.index[company['Date'] == (str) (month) + "/" + (str) (day) + "/" + (str) (year % 100)][0]
    tbl = company.iloc[index: index + n + 1]
    return stock_n_days.append(tbl, ignore_index = True)
```

```
# Same function as above, but it passes date as string
def stock_report_after_n_days(company, n, date):
    stock_n_days = pd.DataFrame()
    index = company.index[company['Date'] == date][0]
    tbl = company.iloc[index: index + n + 1]
    return stock_n_days.append(tbl, ignore_index = True)
```

```
# Calculate change of stock price n days after a particular day in percentage
def price_change(company, n, year, month, day):
    stock = stock_report_after_n_days(company, n, year, month, day)
    return ((stock.iloc[n, 4] - stock.iloc[0, 4]) / stock.iloc[0, 4]) * 100
```

# A/B Testing

Null hypotheses: Elon Musk's tweets about EV do not affect the stock price of EV companies. Any observed differences due to chance

Alternative hypothesis: Elon Musk's optimistic tweets about EV increases the stock price of EV companies.

2 categories for A/B testing: tweets about EV vs tweets not about EV

```
test = tweets.join(tesla_filtered.set_index('Date'), on = 'date').dropna()
test = test[['date', 'TweetsNoPunctuation', 'Mentioned_EV', 'High_low_avg']]
test['Tesla_stock_change'] = [price_change(tesla_filtered, 1, date) for date in test['date']]
test = test[['Mentioned_EV', 'Tesla_stock_change']]
test
```

```
true_avg = np.average(test.loc[test['Mentioned_EV'] == True]['Tesla_stock_change'])
true_avg
```

```
false_avg = np.average(test.loc[test['Mentioned_EV'] == False]['Tesla_stock_change'])
false_avg
```

```
shuffled = test.reindex(np.random.permutation(test.index))
test['Shuffled_labels']= shuffled['Mentioned_EV']
test
```

```
shuffled_true_avg = np.average(shuffled.loc[shuffled['Shuffled_labels'] == True]['Tesla_stock_change'])
shuffled_true_avg
```

```
shuffled_false_avg = np.average(shuffled.loc[shuffled['Shuffled_labels'] == False]['Tesla_stock_change'])
shuffled_false_avg
```

```
true_difference = true_avg - shuffled_true_avg
false_difference = false_avg - shuffled_false_avg
false_difference
```

# Bootstrapping

```
result = []

for _ in range(100):
    shuffled = test.reindex(np.random.permutation(test.index))
    test['Shuffled_labels'] = shuffled['Mentioned_EV']
    shuffled_true_avg = np.average(shuffled.loc[shuffled['Shuffled_labels'] == True]['Tesla_stock_change'])
    true_difference = true_avg - shuffled_true_avg
    result = np.append(result, true_difference)

result
```

```
sum(result > 0)
```

```
sum(result > 0) / len(result)
```
